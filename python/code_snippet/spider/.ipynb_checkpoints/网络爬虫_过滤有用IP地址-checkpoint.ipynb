```python
# 观察者模式
class Subject(object):
    def __init__(self):
        self._observers = []
        
    def attach(self, observer):
        if not observer in self.observers:
            self._observers.append(observer)
            
    def detach(self, observer):
        try:
            self._observers.remove(observer)
        except ValueError as e:
            print('exception: %s' % e)
            
    def notify(self, modifer=None):
        for observer in self._observers:
            if modidier == observer:
                observer.update(self)
                
# 观察者类
class Data(Subject):
    def __init__(self, name=''):
        super(Data, self).__init__()
        self.name = name
        self._data = 0
    
    @property
    def data(self):
        return self._data
    
    @data.setter
    def data(self, value):
        self._data = value
        self.notify()
        
        
```

```python
class Topic(object):
    """主题类。保存所有观察者实例的引用，每个主题都可以有很多观察者
    可以增加和删除观察者"""
    obs = []

    def Attach(self, ob):
        self.obs.append(ob)

    def Detach(self, ob):
        self.obs.remove(ob)

    def Notify(self):
        for ob in self.obs:
            ob.Update()

class Observer(object):
    """抽象观察者类，收到主题的变更通知时，更新自己"""
    def Update(self):
        raise NotImplementedError()      

class ConcreteTopic(Topic):
    """一个具体主题"""
    def __init__(self):
        self.state = None

    def ChangeState(self, newState):
        self.state = newState
        self.Notify()

class ConcreteObserver(Observer):
    """一个具体监听类"""
    def __init__(self, topic):
        self.topic = topic

    def Update(self):
        print("aaa" + self.topic.state)

def client():
    topic = ConcreteTopic()
    ob = ConcreteObserver(topic)
    topic.Attach(ob)
    topic.ChangeState('New State')
    
    
    
client()
```
